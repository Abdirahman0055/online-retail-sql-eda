```python

df = raw_data.rename(columns={
    "InvoiceNo": "order_number",
    "StockCode": "product_code",
    "Description": "product_name",
    "Quantity": "items_sold",
    "InvoiceDate": "order_date",
    "UnitPrice": "price_per_item",
    "CustomerID": "customer_id",
    "Country": "customer_country"
})
```
